```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from glob import glob
```

```python
game_files = glob('Games CSV/ Games Data*.csv')
dataframe_list = [pd.read_csv(game_file) for game_file in game_files]
```

```python
fullgamedata_df = pd.concat(dataframe_list) #adding all csv data to one df.. NOT MERGING
```

```python
fullgamedata_df.head(27)
```

```python
#DataFrame 1 - just grouping by game, grabbing means, and sorting by game rank

game_groupby = fullgamedata_df.groupby(['Game']) #group by game
#get the means for each game, sort by game, rename columns
averages = game_groupby.mean().sort_values("Game Rank").rename(columns = {'Total Viewers':'Average Viewer Count',
                                                                         'Total Channels':'Average Channel Count'})
averages = averages.reset_index(level="Game") #reset index
averages.head()
```

```python
#DataFrame 2 - get max, mean, std for rank, viewers, and channels

no_giantbomb = fullgamedata_df.drop(["Giantbomb ID"], axis = 1) #dropping giantbomb id column, not needed since giantbomb id is static
noGB_game_groupby = no_giantbomb.groupby(['Game']) 

stats = noGB_game_groupby.agg([np.max, np.mean, np.std]) #aggregation with max, mean, and standard dev for rank, viewers, and channels
stats = stats.reset_index(level="Game")
#stats['Total Viewers'] = stats['Total Viewers'].astype(np.float64) ... ignore this
stats = pd.DataFrame(stats)
stats.head(25)

#stats.dtypes
```

```python
#DataFrame 3 - get only the games that have > 27000 viewers. this will help identify specific points in the scatterplots

sort_viewers = stats.sort_values(('Total Viewers','mean'), ascending=False)
#dropcolumns = sort_viewers.drop(["Game Rank"], axis = 1, level=0)
top_viewers = sort_viewers.loc[sort_viewers['Total Viewers']['mean'] > 27000]
top_viewers = top_viewers.reset_index(level=0)
top_viewers.head(50)
```

```python
# Stacked bar graph.. not very helpful

N = len(stats["Game"])
viewers_means = stats["Total Viewers"]["mean"]
channels_means = stats["Total Channels"]["mean"]
ind = np.arange(N)    # the x locations for the groups
width = 0.35       # the width of the bars: can also be len(x) sequence

p1 = plt.bar(ind, channels_means, width)
p2 = plt.bar(ind, viewers_means, width, bottom=channels_means)

plt.ylabel('Total')
plt.title('Streamers to Viewers Ratio per Game')
plt.xticks(ind, (stats["Game"]), rotation='vertical')

plt.yticks(np.arange(0, 180000, 20000))
plt.legend((p1[0], p2[0]), ('Streamers', 'Viewers'))

fig = plt.gcf()
fig.set_size_inches(18.5, 10.5)

plt.show()
```

```python
#Scatterplot 1 - all data

x_limit = 2500
x_axis = np.arange(0, x_limit, 10)
games = stats["Game"]
plt.scatter(channels_means, viewers_means, marker="o", facecolors="purple", edgecolors="black")
plt.xlabel("Streamers")
plt.ylabel("Viewers")
plt.title("Streamers to Viewers Ratio per Game")
fig = plt.gcf()
plt.tight_layout()
fig.savefig("SvV per Game All.png", dpi= 1000)
```

```python
#Scatterplot 2 - games that have less than 1800 streamers, zooming in on clumped group in scatterplot 1

plt.xlim(0, 1800)
plt.scatter(channels_means, viewers_means, marker="o", facecolors="purple", edgecolors="black")
plt.xlabel("Streamers")
plt.ylabel("Viewers")
plt.title("Streamers to Viewers Ratio per Game")
fig = plt.gcf()
plt.tight_layout()
fig.savefig("SvV per Game 0to1800 Streamers.png", dpi= 1000)
```
