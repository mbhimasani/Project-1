```python
import os
import pandas as pd
from glob import glob
```

```python
filenames = glob('Games CSV/ Games Data*.csv')
dataframe_list = [pd.read_csv(f) for f in filenames]
```

```python
all_data_df = pd.concat(dataframe_list)
```

```python
all_data_df
```

```python
all_data_by_game = all_data_df.sort_values(['Game Rank','Game','Total Viewers'])
```

```python
all_data_by_game
```

```python
#Data Analysis
datagroup = all_data_df.groupby('Game')
data_means_df = datagroup.mean()
data_means_df.sort_values("Game Rank")
```

```python
#Counting Frequency of Ranking in First 
datagroup = all_data_df.groupby('Game')
data_means_df = datagroup.mean()
data_means_df.sort_values("Game Rank")
```
